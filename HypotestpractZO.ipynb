```python
###Splitting the groups
chem = df[df['Fertilizer_Type'] == 'chemical']['Growth_Milestone']
org = df[df['Fertilizer_Type'] == 'organic']['Growth_Milestone']
```

```python
chem.head()
```

```python
org.head()
```

```python
###Run t_test
t_stat, p_val = stats.ttest_ind(chem, org)

print(f" 🌱t-statistic: {t_stat:.4}")
print(f" 🌱p-value: {p_val:.4}")
```

```python
alpha = 0.05

if p_val < alpha:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```
